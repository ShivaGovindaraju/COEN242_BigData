```python
import numpy as np
import matplotlib.pyplot as plt

# Author: Shiva K Govindaraju
# Course: COEN 242 Spring 2020
# This is only for generating plots to visualize my data for use in my report
# This does not have any of the implenmentation - simply records of my *many* tests.

shardcnt = [10, 25, 50, 75, 100, 200, 300, 400, 500]
k_10 = [160.06356763839722, 135.77702569961548, 131.8259994983673, 128.627623796463, 130.864506483078, 149.24823927879333, 154.16822218894958, 157.92353558540344, 167.13300323486328]
k_15 = [142.58465433120728, 133.40587711334229, 121.59682583808899, 131.038911819458, 129.67398118972778, 141.55908346176147, 144.87141299247742, 155.38811469078064, 162.87217950820923]
k_25 = [144.56252670288086, 138.30652260780334, 123.13758540153503, 127.94536018371582, 131.26324486732483, 140.06415271759033, 151.5007073879242, 156.1414864063263, 172.8699951171875]

k_15_2 = [151.6818287372589, 133.08566665649414, 126.84074306488037, 129.7062485218048, 132.06539344787598, 140.90089678764343, 153.56432509422302, 159.2979998588562, 169.80627417564392]
k_15_3 = [157.1741693019867, 133.56228184700012, 127.16553592681885, 132.11585450172424, 129.88285946846008, 140.83351039886475, 146.37008619308472, 159.06149697303772, 166.30562162399292]


def plotTotalTimeByK():
    fig, axs = plt.subplots(2, sharey=True)
    axs[0].plot(shardcnt, k_10, 'r-o')
    axs[0].plot(shardcnt, k_15, 'g-o')
    axs[0].plot(shardcnt, k_25, 'b-o')
    axs[0].legend(["k = 10", "k = 15", "k = 25"])

    axs[1].plot(shardcnt[:5], k_10[:5], 'r-o')
    axs[1].plot(shardcnt[:5], k_15[:5], 'g-o')
    axs[1].plot(shardcnt[:5], k_25[:5], 'b-o')
    axs[1].legend(["k = 10", "k = 15", "k = 25"])

    plt.suptitle("Top K Words: Total Execution Time vs Shard Count\n for 400MB Test File")
    axs[0].set(ylabel="Time (Seconds)")
    axs[0].set(xlabel="Shard Count")
    axs[1].set(ylabel="Time (Seconds)")
    axs[1].set(xlabel="Shard Count")
    plt.yticks(range(115,175,10))
    axs[0].set(xticks=range(0,550,50))
    axs[1].set(xticks=range(10,110,10))

    plt.savefig("ExecutionTimeByKAndShards.png")
    
#plotTotalTimeByK()

def plotTotalTimeByKFull():
    plt.plot(shardcnt, k_10, 'r-o')
    plt.plot(shardcnt, k_15, 'b-o')
    plt.plot(shardcnt, k_25, 'g-o')
    plt.legend(["K = 10", "K = 15", "K = 25"], loc="lower right")
    plt.title("Top K Words: Total Execution Time vs Shard Count\n for 400MB Test File")
    plt.ylabel("Total Time (Seconds)")
    plt.xlabel("Shard Count")
    plt.yticks(range(120,180,5))
    plt.xticks(range(0,550,50))
    
    plt.savefig("ExecutionTimeByKAndShards_Full.png")

plotTotalTimeByKFull()
```

```python
def plotTotalTimeByKPart():
    plt.plot(shardcnt[:5], k_10[:5], 'r-o')
    plt.plot(shardcnt[:5], k_15[:5], 'b-o')
    plt.plot(shardcnt[:5], k_15_2[:5], 'c-o')
    plt.plot(shardcnt[:5], k_15_3[:5], 'k-o')
    plt.plot(shardcnt[:5], k_25[:5], 'g-o')
    plt.legend(["K = 10", "K = 15 (1)", "K = 15 (2)", "K = 15 (3)", "K = 25"])
    plt.title("Top K Words: Total Execution Time vs Shard Count\n for 400MB Test File (10-100 Shards only)")
    plt.ylabel("Total Time (Seconds)")
    plt.xlabel("Shard Count")
    plt.yticks(range(120,165,5))
    plt.xticks(range(10,110,10))
    
    plt.savefig("ExecutionTimeByKAndShards_Part.png")

plotTotalTimeByKPart()
```
